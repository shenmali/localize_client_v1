```python
# 💾 CSV'Yİ YÜKLE
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
csv_path = "/content/drive/MyDrive/localization_data/turkish3.csv"  # Colab'a dosya panelinden yüklediğin dosya adı
df = pd.read_csv(csv_path, sep=";")
df = df.dropna(subset=["source", "target"])
df.head()
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.docstore.document import Document

# Embedding modelini tanımla
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Tüm kaynak metinleri belge haline getir
documents = [
    Document(page_content=row["source"], metadata={"translation": row["target"]})
    for _, row in df.iterrows()
]

# FAISS vektör veritabanını oluştur
vector_store = FAISS.from_documents(documents, embeddings)
```

```python
def localize_with_rag(query: str, k: int = 3):
    docs = vector_store.similarity_search(query, k=k)
    print("🔍 En yakın sonuçlar:")
    for i, doc in enumerate(docs, 1):
        print(f"\n{i}. İngilizce: {doc.page_content}\n   Türkçe:   {doc.metadata['translation']}")
```

```python
localize_with_rag("Hello Dr Worry don't worry")
```
